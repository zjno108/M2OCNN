```python
import shutil
import imageio.v2 as imageio
import glob
import os

import imageio.v2 as imageio
import numpy as np
import torch
import torchvision.transforms as transforms
from torch.utils.data import Dataset


train_csv_path = "/home/zhangjian/Repositories/DSLFuse2/csvfiles/train_mri_ct.csv"
test_csv_path = "/home/zhangjian/Repositories/DSLFuse2/csvfiles/test_mri_ct.csv"


mrdir = "/home/zhangjian/Repositories/DSLFuse2/harvard/train/mr"
ctdir = "/home/zhangjian/Repositories/DSLFuse2/harvard/train/ct"
count = 0
for mr_path, ct_path in np.loadtxt(train_csv_path, dtype=str, delimiter=","):
    # 拷贝mr_path文件到mrdir 目录下
    count += 1
    mr = imageio.imread(mr_path)
    ct = imageio.imread(ct_path)
    imageio.imwrite(os.path.join(mrdir, str(count) + ".png"), mr, format="png")
    imageio.imwrite(os.path.join(ctdir, str(count) + ".png"), ct, format="png")

mrdir = "/home/zhangjian/Repositories/DSLFuse2/harvard/val/mr"
ctdir = "/home/zhangjian/Repositories/DSLFuse2/harvard/val/ct"
count = 0
for mr_path, ct_path in np.loadtxt(test_csv_path, dtype=str, delimiter=","):
    # 拷贝mr_path文件到mrdir 目录下
    count += 1
    mr = imageio.imread(mr_path)
    ct = imageio.imread(ct_path)
    imageio.imwrite(os.path.join(mrdir, str(count) + ".png"), mr, format="png")
    imageio.imwrite(os.path.join(ctdir, str(count) + ".png"), ct, format="png")
```
